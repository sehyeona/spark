#  Chapter 01
## 아파치 스파크란

아파치 스파크는 통합 컴퓨팅 엔진이며, 클러스터 환경에서 데이터를 병렬 처리하게 하는 라이브러리 집합이다.  
단일 노트북 부터 수천대 서버로 구성된 클러스터까지 다양한 환경에서 사용이 가능하며, 빅데이터를 다루는 유용한 도구이다.

### 1.1 아파치 스파크의 철학

아파치 스파크 = 빅데이터를 위한 통합 컴퓨팅 엔진과 라이브러리

#### 1) 통합  
빅데이터 애플리케이션에 필요한 통합 컴퓨팅 환경을 제공하자! 라는 목표를 가지고 시작되었다.   
* 그렇다면 통합은 무엇인가??  

스파크는 간단한 데이터 읽기/ SQL처리/ 머신러닝/ 스트림 처리까지 다양한 데이터 분석 작업을 같은 연산엔진과 연관된 API로 수행할 수 있도록 설계되었다. 대부분의 데이터 분석 작업이 다양한 처리 유형과 라이브러리로 나누어져 있기 때문에 불편했음

* 스파크의 통합 특성으로 기존의 데이터 작업을 효율적으로 !!!

먼저 스파크에서는 일관성 있는 조합형 API를 제공  
조합형 API가 부족하면 직접 스파크 기반의 라이브러리 생성 가능 

#### 2) 컴퓨팅 엔진
스파크는 통합이라는 관점을 중시하면서 기능의 범위를 컴퓨팅 엔진으로 제한했다.  
그 결과 스파크는 저장소 시스템의 데이터를 연산하는 역할만 수행할 뿐 영구 저장소 역할을 수행하지 않는다. 
대신 다른 클라우드 기반의 저장소들을 지원한다.  
내부에 데이터를 오래 저장하지 않으며. 특정 저장소를 선호하지도 않는다.
왜냐?? 연구결과 대부분의 데이터가 여러 저장소에 혼합 저장 되어있기 때문 
=> 데이터의 이동은 높은 비용을 초래

* 하둡과의 차이점  
하둡은 저비용장치에 저장하는 __"하둡 파일 시스템"__ 과 맵리듀스 방식의 __"컴퓨팅 시스템"__이 매우 밀접하게 연관되어 
둘중 하나의 시스템만 단독으로 사용하기 어렵다. 
스파크는 하둡 저장소뿐 아니라 다른 모든 저장소와의 호환이 쉽다.

#### 3) 라이브러리
궁극의 스파크 컴포넌트는 데이터 분석 작업에 필요한 통합 API를 제공하는 통합 엔진기반의 자체 라이브러리!!!  
스파크는 표준 라이브러리와 다양한 서드파티 라이브러리를 패키지 형태로 제공  
수백개의 외부 오픈소스 라이브러리가 존재


### 1.2 스파크 등장배경

역사적으로 컴퓨터는 프로세서 성능 향상에 따라 해마다 빨라졌다.   
대부분의 시스템은 따라서 단일 프로세서에서만 실행되도록 설게되었는, 
안타깝게도 하드웨어 단일 프로세서의 성능 향상은 물리적 한계로 인해 2005년경에 멈추었고, 데이터 저장과 수집은 계속 발전한다.   
결과적으로 데이터 수집 비용은 극히 저렴하지만, 데이터는 클러스터에서 처리해야 할 만큼 거대해졌다.   
따라서 새로운 병렬처리 모델들이 필요해졌다

### 1.3 스파크의 역사

UC버클리 스파크 연구 프로젝트에서 탄생하였다.
당시 하둡 맵리듀스는 수천개의 노드로 구성된 클러스터에서 병렬로 데이터를 처리하는 최초의 오픈소스 시스템이자 클러스터 환경용 병렬 프로그래밍 엔진의 대표주자였다.  
스파크 연구 프로젝트팀 AMPLap 은 하둡을 이용한 대규모 머신러닝팀과 함께했고 두가지 사실을 파악한다.

#### 1) 클러스터 컴퓨팅이 엄청난 잠재력을 가지고 있다.  
맵리듀스에 경험이 있는 조직은 기존과는 다른 애플리케이션을 만들어 내었다. 

#### 2) 맵리듀스 엔진을 사용하는 대규모 애플리 케이션의 난이도
머신러닝 팀은 데이터를 10~20회 처리하는데 맵리듀스로 처리하면 단계별로 잡을 개발하고 매번 데이터를 다시 읽어야했다.

이를 해결하기 위해
* 여러 단계로 나누어진 애플리케이션을 간결하게 개발 할 수 있는 함수형 프로그래밍 기반의 API 개발 
* 연산 단계 사이에서 메모리에 저장된 데이터를 효과적으로 읽을 수 있는 새로운 엔진 기반 API 구현

이후 함수형 연산, 구조화된 데이터 등등 효율적인 API들이 계속 추가되고 있다.
현재 DataFrame, 머신러닝 파이프라인, 구조적 스트리밍 등의 강력한 구조체 기반의 신규 API가 추가되고 있다.


### 1.4 스파크의 현재와 미래

예를 들어 2016년에 공개된 구조적 스트리밍은 우버나 넷플릭스 같은 기술회사와 NASA등 연구소에서도 거대한 규모의 데이터 셋을 처리하기 위해 계속 사용된다.  
빅데이터 분석을 수행하는 기업의 핵심기술이 될것이다.
